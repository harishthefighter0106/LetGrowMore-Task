# Exploratory Data Analysis on Terrorism

### Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings("ignore")
```

### LOAD THE DATASETS

```python
df=pd.read_csv("/content/Global Terrorism - START data.zip",encoding='latin1')
df.head()
```

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
df = df[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
df.head()
```

```python
df.shape
```

```python
df.isnull().sum().sum()
```

```python
df['Wound'] = df['Wound'].fillna(0)
df['kill'] = df['kill'].fillna(0)
df['Casualities'] = df['kill'] + df['Wound']
df.info()
```

### Describing the Dataset

```python
df.describe()
```

### Correlation and Heatmap

```python
df.corr()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

### Performing  Data Visualizations

```python
df['Year'].value_counts(dropna=False).sort_index()
```

```python
x_year = df['Year'].unique()
y_count_years = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (15,10))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'pink')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
```

```python
pd.crosstab(df.Year, df.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.xlabel('Year of Attacks')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=df['Country'].value_counts()[:15].index, y=df['Country'].value_counts()[:15].values)
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.subplots(figsize=(10,8))
df.groupby(['Region', 'Year']).count()['Casualities'].plot()
plt.xticks(rotation=45)
plt.ylabel("Casualities")
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.countplot(x='Region', data=df, palette='pink', order=df['Region'].value_counts().index, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_xlabel('Regions')
ax.set_title('Number Of Attacks By Region')
plt.show()
```

```python
Weapon_data = df['Weapon'].value_counts()
Weapon_data
```

```python
plt.figure(figsize= (15,10))
sns.barplot(x = Weapon_data.index , y = Weapon_data.values)
plt.title('Most Used Weapon Type')
plt.xlabel('Weapon Type')
plt.ylabel('Attacks count')
plt.xticks(rotation = 45)
plt.show()
```

```python
df['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color=['green'])
plt.xticks(rotation = 45)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attack")
plt.title("Top 10 Most Effected City")
plt.show()
```

```python
df['Attacktype'].value_counts().plot(kind='bar',figsize=(20,10),color='orange')
plt.xticks(rotation = 45)
plt.xlabel("Attacktype")
plt.ylabel("Number of Attack")
plt.title("Name of AttackType")
plt.show()
```

```python
df[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['indigo'])
plt.xticks(rotation=45)
plt.title("Number of Killed ")
plt.ylabel('Number of People')
plt.xlabel('Attack Type')
plt.show()
```

```python
df['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='pink',figsize=(20,10))
plt.title("Top 10 Terrorist Group Attack")
plt.xlabel("terrorist group name")
plt.xticks(rotation=90)
plt.ylabel("Attack Number")
plt.show()
```

```python
Target_data = df['Targettype'].value_counts().nlargest(n=20)
Target_data
```

```python
plt.figure(figsize= (15,10))
sns.barplot(x = Target_data.index , y = Target_data.values,palette='dark')
plt.title('Most Frequent Target Type')
plt.xlabel('Target Type')
plt.xticks(rotation=45)
plt.ylabel('Attacks count')
plt.xticks(rotation = 90)
plt.show()
```
